```python
# демострируем конкатенацию двух объектов DataFrame с разными столбцами
df1 = pd.DataFrame(np.arange(9).reshape(3, 3), columns=['a', 'b', 'c'])
df2 = pd.DataFrame(np.arange(9, 18).reshape(3, 3), columns=['a', 'c', 'd']) 
df1
```

```python
df2
```

# Слияние и соединение данных

# 1) Слияние данных, расположенных в нескольких объектах

```python
# это наши клиенты
customers = {'CustomersID': [10, 11],
             'Name': ['Mike', 'Marcia'],
             'Address': ['Address for Mike', 'Address for Marcia']}
customers = pd.DataFrame(customers)
customers
```

```python
# это наши заказы
orders = {'CustomersID': [10, 11, 10],
          'OrderDate': [date(2014, 12, 1),
                        date(2014, 12, 1),
                        date(2014, 12, 1)]}
orders = pd.DataFrame(orders)
orders
```

```python
# выполняем слияние датафреймов customers и orders так, чтобы могли отправить товары
customers.merge(orders)
```

```python
# создаем данные, которые будем использовать в качестве примеров в оставшейся части этого раздела
left_data = {'key1': ['a', 'b', 'c'],
             'key2': ['x', 'y', 'z'],
             'lval1': [0, 1, 2]}
right_data = {'key1': ['a', 'b', 'c'],
             'key2': ['x', 'a', 'z'],
             'rval1': [6, 7, 8]}
left = pd.DataFrame(left_data, index=[0, 1, 2])
right = pd.DataFrame(right_data, index=[1, 2, 3])
left
```

```python
right
```

```python
# демострируем слияние, не указывая столбцов, по которым нужно выполнить слияние 
# этот программный код неяно выполняет слияние по всем общим признакам
left.merge(right)
```

```python
# демострирует слияние, явно задав столбец,
# по значениям которого нужно связать объекты DataFrame
left.merge(right, on='key1')
```

```python
# явно выполняем слияние с помощью двух столбцов
left.merge(right, on=['key1', 'key2'])
```

```python
# соединяем индексы строк обеих матриц
pd.merge(left, right, left_index=True, right_index=True)
```

# 2) Настройка семантики соединения при выполнении слияния

```python
# внешнее соединение возвращает все строки из внутреннего соединения,
# а также строки датафреймов left и right, не попавшие во внутреннее соединение
# отсутствующие элементы заполняются NaN
left.merge(right, how='outer')
```

```python
# левое соединение возвращает все строки из внутреннего соединения,
# а также строки датафрейма left, не попавшие во внутреннее соединение
# отсутствующие элементы заполняются NaN
# Итоговый датафрейм содержит общие строки датафрейма left и right с меткой 0 и 2
# а также уникальную строку датафрейма left с меткой 1
# уникальная строка датафрейма left в итоговом датафрейме в столбце rval1 заполняется 
# значением NaN, потому что в датафрейме left этот столбец отсутствовал
left.merge(right, how='left')
```

```python
# правое соединение возвращает все строки из внутреннего соединения,
# а также строки датафрейма right, не попавшие во внутреннее соединение
# отсутствующие элементы заполняются NaN
# Итоговый датафрейм содержит общие строки датафрейма left и right с меткой 0 и 1
# а также уникальную строку датафрейма right с меткой 2
# уникальная строка датафрейма left в итоговом датафрейме в столбце lval1 заполняется 
# значением NaN, потому что в датафрейме right этот столбец отсутствовал
left.merge(right, how='right')
```

```python
# соединяем Left с right, и поскольку эти объекты имеют дублирующиеся имена
# задаем параметры lsuffix и rsuffix
left.join(right, lsuffix='_left', rsuffix='_right')
```

```python
# соединяем left с right c помощью внутреннего соединения
left.join(right, lsuffix='_left', rsuffix='_right', how='inner')
```

# Состыковка и расстановка

# 1) Состыковка с помощью неиерархических индексов

```python
df = pd.DataFrame({'a': [1, 2]}, index=['one', 'two'])
df
```

```python
# помещаем стобец в еще один уровень индекса строк, результатом становится объект Series,
# в котором значения можно просмотреть с помощью мультииндекса
stacked1 = df.stack()
stacked1
```

```python
# ищем значение для 'one'/ 'a', передав кортеж в индексатор
stacked1[('one', 'a')]
```

```python
# создаем датафрейм с двумя столбцами
df = pd.DataFrame({'a': [1, 2],
                   'b': [3, 4]},
                   index=['one', 'two'])
df
```

```python
# помещаем оба столбца в отдельный уровень индекса
stacked2 = df.stack()
stacked2
```

```python
# Ищем значение с помощью индекса 'one' / 'b'
stacked2[('one', 'b')]
```

# 2) Расстыковка с помощью иерархичных индексов

```python
# создаем две копии данных акселерометра, по одной для каждого пользователя
user1 = sensor_readings.copy()
user2 = sensor_readings.copy()
# добавляем столбец who в каждую копию
user1['who'] = 'Mike'
user2['who'] = 'Mikael'
# отмасштабируем данные user2
user2['reading'] *= 100
# и организуем данные так, чтоб получить иерархический индекс строк
multi_user_sensor_data = pd.concat([user1, user2]).set_index(['who', 'interval', 'axis'])
multi_user_sensor_data
```

```python
# извлекаем показания, относящиеся к пользователю Mike, с помощью индекса
multi_user_sensor_data.loc['Mike']
```

```python
# извлекаем все показания по всем осям и по всем пользователям в интервале 1
multi_user_sensor_data.xs(1, level='interval')
```

```python
# выполняем расстыковку, в результате самый внутренний
#  уровень индекса строк (уровень axis) стал уровнем индекса столбцов
multi_user_sensor_data.unstack()
```

```python
# выпоняем расстыковкупо уровню 0
multi_user_sensor_data.unstack(level=0)
```

```python
# выпоняем расстыковку уровней who и axis
unstacked = multi_user_sensor_data.unstack(['who', 'axis'])
unstacked
```

```python
# и, конечно, мы можем выполнить состыковку уровней, которые расстыковали
# выполняем состыковку уровня who
unstacked.stack(level='who')
```
